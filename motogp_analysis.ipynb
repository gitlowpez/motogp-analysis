## Moto GP World Championship(1949-2022)

## Passo 1: Importar as bibliotecas
É necessário importar as bibliotecas de python para o código executar.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Passo 2: Carregar os datasets

Foram importados 6 datasets que correspondem a diferentes dados.

```python
# URLs dos datasets no GitHub
urls = {
    "constructors": "https://raw.githubusercontent.com/dipedilans/assets/refs/heads/main/Moto_GP_World_Championship_1949-2022/constructure-world-championship.csv",
    "gp_events": "https://raw.githubusercontent.com/dipedilans/assets/refs/heads/main/Moto_GP_World_Championship_1949-2022/grand-prix-events-held.csv",
    "gp_winners": "https://raw.githubusercontent.com/dipedilans/assets/refs/heads/main/Moto_GP_World_Championship_1949-2022/grand-prix-race-winners.csv",
    "riders_positions": "https://raw.githubusercontent.com/dipedilans/assets/refs/heads/main/Moto_GP_World_Championship_1949-2022/riders-finishing-positions.csv",
    "riders_info": "https://raw.githubusercontent.com/dipedilans/assets/refs/heads/main/Moto_GP_World_Championship_1949-2022/riders-info.csv",
    "same_nation_podiums": "https://raw.githubusercontent.com/dipedilans/assets/refs/heads/main/Moto_GP_World_Championship_1949-2022/same-nation-podium-lockouts.csv"
}

# Importar os datasets para dataframes do Pandas
df_constructors = pd.read_csv(urls["constructors"])
df_gp_events = pd.read_csv(urls["gp_events"])
df_gp_winners = pd.read_csv(urls["gp_winners"])
df_riders_positions = pd.read_csv(urls["riders_positions"])
df_riders_info = pd.read_csv(urls["riders_info"])
df_same_nation_podiums = pd.read_csv(urls["same_nation_podiums"])


```

## Passo 3: Exploração Inicial dos Dados (df_constructors)




```python
# 1. Explorar a estrutura dos datasets

# Função para visualizar rapidamente as primeiras linhas, tamanho e tipos de dados
def explorar_dataframe(nome, df):
    print(f"\n📌 Estrutura do dataset: {nome}")
    print("-" * 50)
    print("🔹 Primeiras 5 linhas:")
    display(df.head())  # Para melhor visualização no Colab
    print("\n🔹 Dimensões (linhas, colunas):", df.shape)
    print("\n🔹 Informações sobre os dados:")
    print(df.info())
    print("-" * 50)

# Aplicar a função a cada dataset
explorar_dataframe("🏆 Construtores Campeões (df_constructors)", df_constructors)
explorar_dataframe("🛣️ Eventos de GP (df_gp_events)", df_gp_events)
explorar_dataframe("🏁 Vencedores de GP (df_gp_winners)", df_gp_winners)
explorar_dataframe("🏍️ Posições Finais dos Pilotos (df_riders_positions)", df_riders_positions)
explorar_dataframe("📊 Informações sobre Pilotos (df_riders_info)", df_riders_info)
explorar_dataframe("🇳🇦 Pódios com Pilotos da Mesma Nação (df_same_nation_podiums)", df_same_nation_podiums)
```
